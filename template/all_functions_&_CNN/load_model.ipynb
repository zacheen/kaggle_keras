```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
# 準備 data
test_data = pd.read_csv('./input/test.csv')
test_pic = test_data.iloc[:,:].values.reshape(-1,28,28,1)
```

# 方法1 讀取 model

```python
model = tf.keras.models.load_model('./saved_model/final_model')
model.summary()
```

# 方法2 讀取 checkpoint

```python
# 準備 model
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
```

```python
checkpoint_path = "./training_ckpt/cp.ckpt"
checkpoint_dir = "./training_ckpt"
```

```python
# 查看有什麼紀錄
# import os
# os.listdir(checkpoint_dir)
```

```python
# 使用 最新的 check point
latest = tf.train.latest_checkpoint(checkpoint_dir)
print(latest)
model.load_weights(latest)

## 或是 ###########################################

# 使用指定的路徑 
model.load_weights(checkpoint_path)
```

# 測試

```python
# 辨識結果
y_pred = model.predict(test_pic)
y_pred = np.argmax(y_pred, axis=1)
# 把結果輸出到 csv
submission_df = pd.DataFrame()
image_id = [i for i in range(1, 28001)]
submission_df['ImageId'] = image_id
submission_df['Label'] = y_pred
print(submission_df)
submission_df.to_csv('submission.csv', index = False)
```
